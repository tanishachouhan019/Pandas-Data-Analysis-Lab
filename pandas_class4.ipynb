# 🐼 Pandas - Class 4: Data Selection & Filtering
Welcome to **Class 4** of our Pandas series. Today we’ll explore techniques to select and filter data in a DataFrame.

## 1. Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

```python
import pandas as pd
a ={
    'Name':['Jai Rathore', 'Princi', 'Gaurav', 'Anuj', 'Dev'],
    'Age':[27, 24, 22, 32, 33],
    'Score':[80,92,79,85,95],
}
b = pd.DataFrame(a)
print(b)

#1. Select single column Age
print(b[b['Age'] > 27])

#2. Selcting single column score > 85
print(b[b['Score']> 85])

#3. Multiple conditions (AND): Age > 27 AND(&) Score > 85
print(b[(b['Age'] > 27) & (b['Score'] > 85)])

#4. Multiple conditions (OR): Age > 27 OR(|) Score > 85
print(b[(b['Age'] > 27) | (b['Score'] > 85)])

#5. Multiple conditions (NOT): Age > 27 NOT(~) Score > 85
print(b[~(b['Score'] > 85)])

#6. Age > 22 AND Score > 85
print(b[(b['Age'] > 22) & (b['Score'] > 85)])
```

## 2. Filtering Rows with Multiple Conditions
- Combine two or more conditions to refine selection.
- Remember to enclose each condition in parentheses.

```python
#1. Age > 25 AND Score > 85
print(b[(b['Age'] > 25) & (b['Score'] > 85)])

#2. Age < 35 or score >= 90
print(b[(b['Age'] < 35) | (b['Score'] >= 90)])

#3. Age between 25 and 35 (inculsive)
print(b[(b['Age'] >= 25) & (b['Age'] <= 35)])
```

## 3. Using `query()` for Cleaner Filtering
- `query()` allows filtering rows with a string expression.
- It's concise and readable for complex conditions.
- Variables can be passed using the `@` symbol.

```python
#1. Simple Query ; Age > 25
print(b.query('Age > 25'))

#2. Multiple conditions using query: Age <35 andScore >= 85
print(b.query("Age < 35 and Score >=85"))

#3. Multiple Conditions using query : Age >= 30 and Score >=90
print(b.query("Age >= 30 and Score >=90"))

#4. Threshold (@)
score = 90 #Variable
print(b.query("Score >= @score"))
```

## 4. String Operations on Columns
- Use the `.str` accessor for vectorized string methods.
- Common methods: `contains`, `replace`, `startswith`, `endswith`, `lower`, `upper`.
- Useful for cleaning or filtering text data.

```python
print(b)
#1. Convert nsme to lowercase()
print(b['Name'].str.lower())

#2. Convert name to uppercase()
print(b['Name'].str.upper())

#3. title() -> Captialize first letter of all the words
print(b['Name'].str.title())

#4. Replace a substring('OldString','NewString')
print(b['Name'].str.replace('Charlie','Charles'))

#5. Check if Name column contains the letter "a"
print(b[b['Name'].str.contains('a')])

#6. .Startswith('Letter') -> Letter staring with.
print(b[b['Name'].str.startswith('D')])

#7. .endswith('letter') -> letter Ending with.
print(b[b['Name'].str.endswith('e')])
```

## Mini Practice
1. Build a DataFrame with columns such as Name, Age, City, Score.
2. Use boolean indexing to select rows where Age > 25.
3. Filter rows where City is either 'Delhi' or 'Mumbai'.
4. Use query() to find rows with Score >= 85 and Age < 35.
5. Clean a text column using str.replace or str.lower.

```python


# 1. Build a DataFrame with columns: Name, Age, City, Score
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 22, 35, 28],
    'City': ['New York', 'London', 'Paris', 'Berlin', 'Sydney'],
    'Score': [85, 92, 78, 95, 88]
}

df = pd.DataFrame(data)
print(df)


# 2. Use boolean indexing to select rows where Age > 25
print(df[df['Age'] > 25])

# 3. Filter rows where City is either 'Delhi' or 'Mumbai'
print(df[df['City'].isin(['Delhi', 'Mumbai'])])

# 4. Use query() to find rows with Score >= 85 and Age < 35
print(df.query('Score >= 85 and Age < 35'))

# 5. Clean a text column using str.replace or str.lower
# Example: convert City names to lowercase or replace "delhi" with "New Delhi"
print(df['City'].str.lower())
print(df['City'].str.replace('delhi', 'New Delhi'))
```
