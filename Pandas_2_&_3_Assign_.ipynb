# Q1. Display First Rows

## Problem Statement
Create the following DataFrame:

| Name | Age |
|------|-----|
| Aman | 21 |
| Rahul | 22 |
| Neha | 20 |
| Priya | 23 |
| Ravi | 24 |

Display the first 3 rows using `head()`.

## Sample Output
```python
First 3 rows of DataFrame
```


````python
import pandas as pd
a = {
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 24]
}

b = pd.DataFrame(a)

print(b)
print(b.head(3))
````


# Q5. Select Data using loc

## Problem Statement
Create a DataFrame with Name, Age and City.

Using `loc`, display the Name and City of the first row.

## Sample Output
```python
Name    Aman
City    Bhopal
```


````python
import pandas as pd
a = {
    "Name": ["Aman", "Rahul", "Neha"],
    "Age": [21, 22, 20],
    "City": ["Bhopal", "Indore", "Bhopal"]
}
b = pd.DataFrame(a)
print(b.loc[0, ["Name", "City"]])
````


# Q6. Select Data using iloc

## Problem Statement
Create a DataFrame with at least 4 rows.

Using `iloc`, print:
- Second row
- First two columns

## Sample Output
```python
Selected row values
```


````python
import pandas as pd
a = {
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Age": [21, 22, 20, 23],
    "City": ["Bhopal", "Indore", "Bhopal", "Indore"]
}
b = pd.DataFrame(a)
print(b.iloc[1])
print(b.iloc[0:2, 0:2])
````


# Q7. Rename a Column

## Problem Statement
Create a DataFrame:

```python
Name
Marks
```

Rename column `Marks` to `Score`.

## Sample Output
```python
Name  Score
```


````python
import pandas as pd
a = {
    "Name": ["Aman", "Rahul", "Neha"],
    "Marks": [80, 75, 90]
}
b = pd.DataFrame(a)
b.rename(columns={"Marks": "Score"}, inplace=True)
print(b)
````


# Q8. Handle Missing Values


# Q10. Remove Duplicate Records

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman","Rahul","Aman","Neha"]
```

Check duplicate rows using `duplicated()` and remove them using `drop_duplicates()`.

## Sample Output
```python
Aman
Rahul
Neha
```


````python
import pandas as pd
a = {
    "Name": ["Aman", "Rahul", "Aman", "Neha"]
}
b = pd.DataFrame(a)
b.drop_duplicates(inplace=True)
print(b)
````
